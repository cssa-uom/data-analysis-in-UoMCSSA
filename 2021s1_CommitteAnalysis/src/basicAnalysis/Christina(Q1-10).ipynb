```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties

font = FontProperties(fname='/System/Library/Fonts/STHeiti Light.ttc')
plt.rcParams['font.size'] = '15'
```

```python
data = pd.read_csv('data.csv')
```

```python
# 性别
sex = ["male", "female", "other"]
value = [53, 87, 1]

fig, ax = plt.subplots(figsize=(20,10), subplot_kw=dict(aspect="equal"))

sex = [i for i in sex]
value = [int(num) for num in value]

def func(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d})".format(pct, absolute)

wedges, texts, autotexts = ax.pie(value, autopct=lambda pct: func(pct, value),
                                  textprops=dict(color="w"))

ax.legend(wedges, sex,
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1),
          fontsize = 15)

plt.setp(autotexts, size=12, weight="bold")
ax.set_title("CSSA成员性别比例0", fontproperties=font, fontsize=20)
```

```python
# 生日月份
month = ["January", "Ferbuary", "March", "April", "May", "June",
        "July", "August", "September", "October", "November", "December"]
count_m = [13, 9, 9, 14, 11, 13, 15, 15, 9, 10, 10, 13]

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')


plt.figure(figsize = (20, 10))
plt.bar(month, count_m)
addlabels(month, count_m)
plt.xlabel('Month',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title("CSSA每月生日的人数",fontproperties=font, fontsize=20)
plt.show()
```

```python
#星座
constellations= ["Aries", "Taurus", "Gemini", "Cancer", "Leo", "Virgo",
    "Libra", "Scorpio", "Sagittarius", "Capricorn", "Aquarius", "Pisces"]
count_c = [14, 12, 15, 14, 15, 9, 11, 9, 10, 15, 10, 7]

plt.figure(figsize = (20, 10))
plt.bar(constellations, count_c)
addlabels(constellations, count_c)
plt.xlabel('Constellations',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title("CSSA内相同星座的人数", fontproperties=font, fontsize=20)
plt.show()
```

```python
# 地区分布
province = data.iloc[:,3].value_counts().index.tolist()
count_p = data.iloc[:,3].value_counts().tolist()

data2 = {'province':province, 'count':count_p}
df2 = pd.DataFrame(data2)

df2.drop(df2[df2['province'] == '真不知道'].index, inplace = True)
df2.drop(df2[df2['province'] == '保密'].index, inplace = True)
df2.drop(df2[df2['province'] == '中国'].index, inplace = True)

province = list(df2['province'])
count_p = list(df2['count'])

plt.figure(figsize = (20, 10))
plt.bar(province, count_p)
addlabels(province, count_p)
plt.xticks(np.arange(28),province,fontproperties=font,rotation=90,size=15)
plt.xlabel('Province',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title("CSSA内相同家乡的人数",fontproperties=font, fontsize=20)
plt.show()
```

```python
# 具体家乡
pAc = data.iloc[:,4].value_counts().index.tolist()
count_p1 = data.iloc[:,4].value_counts().tolist()

data3 = {'省市':pAc, '人数':count_p1}
df3 = pd.DataFrame(data3)

df3.drop(df3[df3['省市'] == '真不知道'].index, inplace = True)
df3.drop(df3[df3['省市'] == '保密'].index, inplace = True)
df3.drop(df3[df3['省市'] == '中国'].index, inplace = True)
df3.drop(df3[df3['省市'] == '浙江省'].index, inplace = True)
df3.drop(df3[df3['省市'] == '江苏省'].index, inplace = True)
df_sort = df3.sort_values(by=['省市'])
df_sort.to_csv('相同家乡.csv',index = False)
```
